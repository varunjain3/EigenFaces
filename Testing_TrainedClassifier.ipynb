```python
# inline for plots to show up in jupyter notebook
%matplotlib inline

# Some dependencies
# Make sure you run the command "pip install -r requirements.txt before"
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import PIL
from tqdm import tqdm
import os
```

```python
face_c = cv.CascadeClassifier("classifier\cascade.xml")

# As working with images, this function plots any binnary image
def show_im(image):
    plt.imshow(image,cmap="gray")
    plt.show()
```

```python
DATASET = os.path.join(os.getcwd(),"YALE") 
print("Dataset is ",DATASET)

count = 0
minArea = 20000
show_image = 5

for dirname,_,filenames in os.walk(DATASET):
    
    for file in filenames:
        if count > show_image:
                break
        im = PIL.Image.open(os.path.join(dirname,file))
        im = np.array(im)
        faces = face_c.detectMultiScale(im,minNeighbors=3)

        for face in faces:            
            face_im = im[face[1]:face[0],face[1] + face[3]:face[0]+face[2]]
            if face_im.shape[0]*face_im.shape[1] > minArea:
                count += 1
                show_im(face_im)
```
